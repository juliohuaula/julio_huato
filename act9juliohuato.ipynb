```python
#b0 + b1*x1
modelA=sm.OLS.from_formula('cadera~peso',data=df)
resultA=modelA.fit()
resultA.summary()
```

```python
print('b0=-8.3137')
print('b1=0.8316')
print('El modelo final sería: peso= -8.3137 + 0.8316*cintura')
```

```python
resi=resultA.resid
resi.median()
```

```python
df.corr()
```

```python
sns.regplot(x='cintura',
            y='peso',
            data=df,
           )
```
