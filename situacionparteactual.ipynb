# Situación problema

Punto número 1: El nombre que caracteriza el conjunto de datos es la Tasa de suicidios de 1985 a 2016. Este conjunto de datos compilado se extrajo de otros cuatro conjuntos de datos vinculados por tiempo y lugar, y se construyó para encontrar señales correlacionadas con el aumento de las tasas de suicidio entre diferentes cohortes a nivel mundial, en todo el espectro socioeconómico.

Punto número 2: Estos datos están confeccionados a raíz de datos obtenidos de organizaciones como la ONU (UN), el Banco Mundial (World Bank) y la Organización Mundial de la salud (WHO).

Punto número 3: El enlace para descargar la base de datos es la siguiente: https://www.kaggle.com/russellyates88/suicide-rates-overview-1985-to-2016 .

Punto número 4: Los datos que contiene son: el país, y año donde aconteció el suceso, además de la edad y sexo de la víctima. Del mismo modo contiene la población total de cada año específico y el índice entre 100 mil personas.
Finalmente tenemos el aspecto socioeconómico donde se tiene el Human Development Index, el Producto Interno Bruto (GDP) y Producto Interno Bruto per cápita y la generación a la que pertenecía el fallecido; ej: Boomer.

Punto número 5: En la base de datos se entienden absolutamente todos los encabezados; no obstante, se tuvo que investigar qué era el HDI porque no se sabía el concepto esencial. Una vez investigado, se pudo llegar a la conclusión de que correspondía a Human Development Index.

Punto número 6: Alguna de las preguntas que se podrían hacer son:

```python
pivcienmil= pd.pivot_table(df,
                           index='year',
                           values= 'suicides/100k pop',
                           columns=['country'],
                           aggfunc= 'mean'
                          )
pivcienmil
```
